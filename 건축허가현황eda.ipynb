```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## 연도별 건축허가현황

* 건축허가 연면적의 용도별 변화현황
* 건축허가 동수의 용도별 변화현황 (ex - 연면적은 좁지만 동수는 많을 수 있음)
* 2011~2020년의 동향


```python
raw_data = pd.read_excel("../project/건축허가현황.xlsx")
raw_data
```

```python
raw_data.columns
```

```python
area_data = raw_data[['연면적-주거용[㎡]', '연면적-상업용[㎡]', '연면적-공업용[㎡]', '연면적-교육및사회용[㎡]','연면적-기타(용도별)[㎡]','연면적-용도별[㎡]']]
building_data = raw_data[['동수-주거용[동수]', '동수-상업용[동수]','동수-공업용[동수]', '동수-교육및사회용[동수]', '동수-기타(용도별)[동수]','동수-용도별[동수]']]
```

```python
area_data = area_data.rename({'연면적-주거용[㎡]':'주거용', '연면적-상업용[㎡]':'상업용', '연면적-공업용[㎡]':'공업용', '연면적-교육및사회용[㎡]':'교육및사회용','연면적-기타(용도별)[㎡]':'기타','연면적-용도별[㎡]':'총 연면적'},axis='columns')
area_data.head()
```

```python
building_data = building_data.rename({'동수-주거용[동수]':'주거용', '동수-상업용[동수]':'상업용','동수-공업용[동수]':'공업용', '동수-교육및사회용[동수]':'교육및사회용', '동수-기타(용도별)[동수]':'기타','동수-용도별[동수]':'총 건물 수'},axis='columns')
building_data.head()
```

```python
area_data.index.dtype  # 타임스탬프
```

```python
area_data.index[0].year
```

```python
year = [] 
for x in area_data.index:
    year.append(x.year)   
```

```python
area_data.index = year
building_data.index = year
```

```python
from matplotlib import font_manager, rc #한글처리
f_path = 'C:/Windows/Fonts/HANDotum.ttf'
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font',family = font_name)
plt.rcParams['axes.unicode_minus'] = False
```

# 2011~2020 용도별 건축허가 연면적

```python
area_data = area_data.groupby(by=area_data.index).sum().loc['2011':'2020']
area_data
```

```python
plt.figure(figsize = (10,10))
plt.bar(area_data.index, '총 연면적',data= area_data)
plt.title('용도별 총 연면적 연도별 변화')
plt.xlabel('연도')
plt.ylabel('총 연면적')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(area_data.index,area_data['주거용'],label='주거용')
plt.plot(area_data.index,area_data['상업용'],label='상업용')
plt.plot(area_data.index,area_data['공업용'],label='공업용')
plt.plot(area_data.index,area_data['교육및사회용'],label='교육및사회용')
plt.plot(area_data.index,area_data['기타'],label='기타')

plt.xlabel('연도')
plt.ylabel('용도별 건축허가 연면적')
plt.legend(loc='upper right')
plt.show()
```

# 2011~2020 용도별 건축허가 건물 수

```python
building_data = building_data.groupby(by=building_data.index).sum().loc['2011':'2020']
building_data
```

```python
plt.figure(figsize = (10,10))
plt.bar(building_data.index, '총 건물 수',data= building_data)
plt.title('용도별 총 건물 수 연도별 변화')
plt.xlabel('연도')
plt.ylabel('총 건물 수')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(building_data.index,building_data['주거용'],label='주거용')
plt.plot(building_data.index,building_data['상업용'],label='상업용')
plt.plot(building_data.index,building_data['공업용'],label='공업용')
plt.plot(building_data.index,building_data['교육및사회용'],label='교육및사회용')
plt.plot(building_data.index,building_data['기타'],label='기타')

plt.xlabel('연도')
plt.ylabel('용도별 건축허가 건물 수')
plt.legend(loc='upper right')
plt.show()
```
