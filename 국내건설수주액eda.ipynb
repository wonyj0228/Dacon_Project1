# 연도별 국내 건설수주액

**건설수주액이란?**<br>
일반적으로 건설업체가 발주자와 체결한 공사계약액을 말한다.

```python
raw_data = pd.read_excel('../project/국내건설수주액.xlsx')
raw_data
```

```python
raw_data=raw_data[['공공부문[백만원]','민간부문[백만원]']]
raw_data
```

```python
raw_data.index.dtype  #타임스탬프
```

```python
year = [] 
for x in raw_data.index:
    year.append(x.year)  
raw_data.index = year
```

```python
raw_data=raw_data.groupby(by=raw_data.index).sum()
raw_data=raw_data.loc['2011':'2020']
raw_data
```

```python
from matplotlib import font_manager, rc #한글처리
f_path = 'C:/Windows/Fonts/HANDotum.ttf'
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font',family = font_name)
plt.rcParams['axes.unicode_minus'] = False
```

```python
plt.figure(figsize=(10,10))
plt.bar(raw_data.index,raw_data['공공부문[백만원]'])
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.bar(raw_data.index,raw_data['민간부문[백만원]'])
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.plot(raw_data.index,raw_data['공공부문[백만원]'],label='공공부문')
plt.plot(raw_data.index,raw_data['민간부문[백만원]'],label='민간부문')
plt.legend()
plt.show()
```
