```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_data = pd.read_excel('../project/주택건설인허가실적.xlsx')
raw_data
```

```python
raw_data.index.dtype
```

```python
year=[]
for x in raw_data.index:
    year.append(x.year)

raw_data.index = year
raw_data
```

```python
raw_data.columns
```

```python
area_data=raw_data[['부산[호]', '충북[호]', '충남[호]', '대구[호]', '대전[호]',
          '강원[호]', '광주[호]','경기[호]', '인천[호]', '제주[호]', 
          '전북[호]', '전남[호]', '경북[호]', '경남[호]','세종[호]', 
          '서울[호]', '울산[호]']]
area_data
```

```python
area_data=area_data.groupby(by=area_data.index).sum().loc['2011':'2020']
area_data
```

```python
from matplotlib import font_manager, rc #한글처리
f_path = 'C:/Windows/Fonts/HANDotum.ttf'
font_name = font_manager.FontProperties(fname=f_path).get_name()
rc('font',family = font_name)
plt.rcParams['axes.unicode_minus'] = False
```

# 행정구역 8도

```python
area8 =area_data[
        ['충북[호]','충남[호]','강원[호]','경기[호]',
        '전북[호]','전남[호]','경북[호]','경남[호]']
         ]
area8
```

```python
plt.figure(figsize=(10,10))
plt.plot(area8.index,area8['경기[호]'],label='경기')
plt.plot(area8.index,area8['강원[호]'],label='강원')
plt.plot(area8.index,area8['충북[호]'],label='충북')
plt.plot(area8.index,area8['충남[호]'],label='충남')
plt.plot(area8.index,area8['전북[호]'],label='전북')
plt.plot(area8.index,area8['전남[호]'],label='전남')
plt.plot(area8.index,area8['경북[호]'],label='경북')
plt.plot(area8.index,area8['경남[호]'],label='경남')

plt.xlabel('연도')
plt.ylabel('행정구역별 주택허가')
plt.legend(loc='upper right')
plt.show()
```

# 특별시, 광역시

```python
city=area_data[['서울[호]','부산[호]','대구[호]','대전[호]','광주[호]',
           '인천[호]','제주[호]','세종[호]','울산[호]']]
city
```

```python
plt.figure(figsize=(10,10))
plt.plot(city.index,city['서울[호]'],label='서울')
plt.plot(city.index,city['부산[호]'],label='부산')
plt.plot(city.index,city['대구[호]'],label='대구')
plt.plot(city.index,city['대전[호]'],label='대전')
plt.plot(city.index,city['광주[호]'],label='광주')
plt.plot(city.index,city['인천[호]'],label='인천')
plt.plot(city.index,city['제주[호]'],label='제주')
plt.plot(city.index,city['세종[호]'],label='세종')
plt.plot(city.index,city['울산[호]'],label='울산')

plt.xlabel('연도')
plt.ylabel('특별/광역시 별 주택허가')
plt.legend(loc='upper right')
plt.show()
```

# 공공부문,민간부문

```python
pp=raw_data[['민간부문[호]','공공부문[호]']]
pp=pp.groupby(by=pp.index).sum().loc['2011':'2020']
pp
```

```python
plt.figure(figsize=(10,10))
plt.plot(pp.index,pp['민간부문[호]'],label='민간')
plt.plot(pp.index,pp['공공부문[호]'],label='공공')


plt.xlabel('연도')
plt.ylabel('민간/공공 별 주택허가')
plt.legend(loc='upper right')
plt.show()
```
