```python
data3 = pd.read_csv("../dataproject/주택매매가격지수(KB).csv", encoding="utf-8", engine="python")
```

```python
data3=data3.iloc[ : ,0:2]
```

```python
data3=data3.rename({'Unnamed: 0' : '일자'},axis='columns')
```

```python
data3['일자'] = data3['일자'].astype(dtype= str)
```

```python
data3['일자'][0][0:7]
```

```python
def date_cleaning(text):
    result = re.sub("-","",text)
    return result
```

```python
data3['일자'] = data3['일자'].apply(lambda x : date_cleaning(x))
```

```python
date = []
for x in data3['일자']:
    date.append(x[0:6])
data3['일자'] = date
```

```python
data3.index = data3['일자']
```

```python
data3=data3.drop(columns='일자')
```

```python
data3 = data3['201101':'202012']
```

```python
data3=data3.rename({'총지수[2019.01=100]' : '총지수'},axis='columns')
```

```python
data3['총지수'] = data3['총지수'] - 0.576
```

```python
data3.to_csv('주택매매지수.csv')
```
